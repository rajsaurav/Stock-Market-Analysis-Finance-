## Tesla, Ford and General Motors Stocks' Market Analysis for 2024

```python
import yfinance as yf
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
start = '2024-01-01'
end = '2024-10-31'

# Fetch data using yfinance
tesla = yf.download('TSLA',start,end)
ford = yf.download('F',start,end)
gm = yf.download('GM',start,end)
```

```python
tesla.head()
```

```python
ford.head()
```

```python
gm.head()
```

## Opening price for different stocks in 2024

```python
tesla['Open'].plot(label = 'Tesla')
plt.ylabel('Tesla Opening Price')
ford['Open'].plot(label = 'ford', color ='Orange')
plt.ylabel('Ford Opening Price')
gm['Open'].plot(label = 'GM',color = 'Green')
plt.ylabel('General Motors Opening Price')
plt.show()
```

## Volume Traded Over Time

```python
stocks = ["TSLA", "F", "GM"]

# Download stock data
stock_data = yf.download(stocks, start=start, end=end)['Volume']

# Plot each stock's data on the same graph
plt.figure(figsize=(14, 7))
for stock in stocks:
    plt.plot(stock_data.index, stock_data[stock], label=stock)

# Add labels and title
plt.xlabel("Date")
plt.ylabel("Volume Traded")
plt.title("Volume Traded Over Time")
plt.legend()  # Show legend for each stock symbol
plt.grid(True)
plt.show()
```

```python
ford['Volume'].idxmax()
```

## Important Conclusion

There was sudden surge in volume trading by ford towards the end of July

```python
stocks = ["TSLA", "F", "GM"]

# Download stock data
stock_data = yf.download(stocks, start=start, end=end)['Adj Close']

# Plot each stock's data on the same graph
plt.figure(figsize=(14, 7))
for stock in stocks:
    plt.plot(stock_data.index, stock_data[stock], label=stock)

# Add labels and title
plt.xlabel("Date")
plt.ylabel("Adjusted Close Price")
plt.title("Stock Prices Over Time")
plt.legend()  # Show legend for each stock symbol
plt.grid(True)
plt.show()
```

## Daily Percentage Change

```python
tesla['returns'] = (tesla['Close']/tesla['Close'].shift(1) -1)
```

```python
ford['returns'] = (ford['Close']/ford['Close'].shift(1) -1)
```

## Box Plots

```python
# Concatenate the 'returns' columns from each DataFrame
box_df = pd.concat([tesla['returns'], ford['returns'], gm['returns']], axis=1)
box_df.columns = ['Tesla Returns','Ford Returns','GM Return']
box_df.plot(kind = 'box',figsize = (16,6))
```

```python
scatter_matrix(box_df,figsize = (8,8),hist_kwds = {'bins':50})
```

## Insights and Recommendations

- Around end of July and beginning of August in 2024 Ford's shares opening price took a major dip and so the trading volume skyrocketed. This was due to following reasons:

        (a) There were deficiencies in the quality assurance of certain vehicle models
        (b) These deficiencies were leading to increased warranty costs
        (c) The Company’s warranty reserves did not accurately reflect the quality issues of the sold vehicles 
        (d) As a consequence, the Company’s profitability was likely to be negatively impacted


- Ford and GM almost share a relationship which can be seen in scatter matrix. This can be because they are not an electric vehicle company like Tesla. 


- From the box plot you can conclude that tesla's stocks are more volatile. 

- In 2024 the ford and GM gave double returns which is not the case with tesla.
